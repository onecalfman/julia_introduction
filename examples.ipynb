### Example: Numerical Differentiation
The following function returns a vector with all differences between a vectors elements.
In the second definition, a timespan can be provided and the values will be scales acordingly.
Note that i doubled the last value in the vector, to make the input and output length equal.

This example is a bit hard to read, but is a good example for using different features of
julia, for composing functions in a very terse way. It also uses the multiple dispatch feature.

```julia
vecdiff(vec::Vector) = map((v1, v2) -> v1 - v2, vec, [vec[2:end]; vec[end]])
vecdiff(vec::Vector, interval::Number) =  vecdiff(vec) ./ (interval ./ length(vec))
```

```julia
vecdiff(vec)
```

```julia
vecdiff(vec, 20)
```

```julia
range = -10:0.1:10
sig   = x -> 1/(1 + exp(-x))
vals  = sig.(range)
plot(range, [vals, vecdiff(vals,5)])
```
